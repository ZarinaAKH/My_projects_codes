```python
import struct
from pathlib import Path

file_path_1 = Path(r"D:/Master`s Degree/2_semester/homework/28.DAT")

with file_path_1.open('rb') as f:
    raw_bytes = f.read()

# Размер одного значения (4 байта)
ELEMENT_SIZE = struct.calcsize('i')  
SKIP_ELEMENTS = 7  # Пропускаем первые 8 элементов (индексация с 0)

# Вычисляем смещение и берем оставшиеся данные
offset = SKIP_ELEMENTS * ELEMENT_SIZE
remaining_bytes = raw_bytes[offset:]

# Используем iter_unpack для обработки данных
unpacked_data_1 = [
    val[0] 
    for val in struct.iter_unpack('i', remaining_bytes)
]

print(unpacked_data_1)
```

```python
import struct
from pathlib import Path

file_path_2 = Path(r"D:/Master`s Degree/2_semester/homework/89.DAT")

with file_path_2.open('rb') as f:
    raw_bytes = f.read()

# Размер одного значения (4 байта)
ELEMENT_SIZE = struct.calcsize('i')  
SKIP_ELEMENTS = 7  # Пропускаем первые 8 элементов (индексация с 0)

# Вычисляем смещение и берем оставшиеся данные
offset = SKIP_ELEMENTS * ELEMENT_SIZE
remaining_bytes = raw_bytes[offset:]

# Используем iter_unpack для обработки данных
unpacked_data_2 = [
    val[0] 
    for val in struct.iter_unpack('i', remaining_bytes)
]

print(unpacked_data_2)
```

```python
import numpy as np
from scipy.signal import find_peaks
import matplotlib.pyplot as plt

# Оптимальные параметры (подбирайте под ваши данные)
params = {
    'height': 0.02,      # Более низкий порог
    'distance': 3,       # Меньшее минимальное расстояние
    'prominence': 0.03,  # Учитывает "выдающиеся" пики
    'width': 1           # Учитывает ширину пиков
}

# Находим пики с новыми параметрами
peaks, properties = find_peaks(processed_spectrum_2, **params)

# Визуализация с информацией о пиках
plt.figure(figsize=(15, 8))
plt.plot(angle_range_2, processed_spectrum_2, label='Сигнал')
plt.scatter(angle_range_2[peaks], processed_spectrum_2[peaks], 
            color='red', s=50, label=f'Найдено пиков: {len(peaks)}')

# Подписи для каждого пика
for i, peak in enumerate(peaks):
    plt.text(angle_range_2[peak], processed_spectrum_2[peak]+0.01, 
             f'{i+1}\nH={processed_spectrum_2[peak]:.2f}', 
             ha='center', fontsize=8)

plt.xlabel('Угол (2θ)', fontsize=12)
plt.ylabel('Интенсивность', fontsize=12)
plt.title(f'Анализ пиков (обнаружено {len(peaks)} пиков)', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```

```python
min_height = 0.05
# Минимальная высота пика
min_distance = 5   # Минимальное расстояние между пиками

# Находим все пики
peaks, _ = find_peaks(processed_spectrum_1, height=min_height, distance=min_distance)

# Удаляем первый пик (если нужно)
peaks = peaks[3:]
plt.figure(figsize=(15,8))
plt.plot(angle_range_1, processed_spectrum_1, label='Сигнал')
plt.plot(angle_range_1[peaks], processed_spectrum_1[peaks], 'ro', label='Пики')
plt.xlabel('Номер измерения')
plt.ylabel('Интенсивность')
plt.title('Пики')

plt.legend()
```

```python
def sliding_section(data, length=40):
    filt_data = data.copy()
    replaced_section = np.arange(1, length-1, 1)
    for i in range(len(filt_data)):
        if i < length:
            continue
        x = (i-length, i)
        y = (filt_data[i-length], filt_data[i])
        coef = np.polyfit(x, y, 1)
        line = np.array([coef[0]*xr+coef[1] for xr in replaced_section])
        repl_mask = line < filt_data[i-length+1:i-1]
        filt_data[i-length+1:i-1][repl_mask] = line[repl_mask]
        replaced_section += 1
    return filt_data

# Применяем функцию к вашим данным
noise_28 = sliding_section(processed_spectrum_1)
noise_89 = sliding_section(processed_spectrum_2)

# Создаем графики
fig, axs = plt.subplots(2, 1, figsize=(15, 10))
axs[0].plot(angle_range_1, processed_spectrum_1, label="28")
axs[1].plot(angle_range_2, processed_spectrum_2, label="89")
axs[0].plot(angle_range_1, noise_28, label="noise")
axs[1].plot(angle_range_2, noise_89, label="noise")

# Настраиваем графики
for ax in axs:
    ax.set_xlabel("Угол")
    ax.set_ylabel("Амплитуда")
    ax.set_title("Дифрактограмма")
    ax.legend()
    ax.grid(True)

fig.tight_layout()
plt.show()
```

```python
# Находим пики для образца 89 с учетом ширины пиков
peaks_89, properties = find_peaks(processed_spectrum_2, height=0.05, distance=10, prominence=0.1, width=3)

# Выбираем центральный пик (второй по счету)
peak_idx = 1

# Автоматически определяем размер окна на основе ширины пика
window_size = int(properties['widths'][peak_idx] * 4)  # Берем окно в 4 раза шире пика
if window_size < 10:  # Минимальный размер окна
    window_size = 10

# Выделяем область импульса с базовой линией
standard_impulse = processed_spectrum_2[peaks_89[peak_idx]-window_size:peaks_89[peak_idx]+window_size] 
standard_noise = noise_89[peaks_89[peak_idx]-window_size:peaks_89[peak_idx]+window_size]
standard_angles = angle_range_2[peaks_89[peak_idx]-window_size:peaks_89[peak_idx]+window_size]

# Нормализуем импульс
standard_impulse = standard_impulse - standard_noise
standard_impulse = standard_impulse - np.min(standard_impulse)  # Сдвигаем минимум к нулю

# Визуализация с улучшениями
plt.figure(figsize=(12, 6))

# Основной график импульса
plt.plot(standard_angles, standard_impulse, 'b-', linewidth=2.5, label='Эталонный импульс')

# Настройки графика
plt.title("Эталонный импульс", fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)


plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
# Вычисляем корреляцию для образца 89
xrd_corr = np.correlate(processed_spectrum_2 - noise_89, standard_impulse, mode='full')[standard_impulse.shape[0]-1:]

# Визуализация корреляции
plt.figure(figsize=(15, 5))
plt.plot(angle_range_2, xrd_corr)
plt.title("Корреляция с эталонным импульсом (образец 89)")
plt.xlabel("Угол")
plt.ylabel("Корреляция")
plt.grid(True)
plt.show()
```
